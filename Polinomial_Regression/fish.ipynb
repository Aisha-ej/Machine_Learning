```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

```python
x = df.iloc[:, [0]] 
x
```

```python
y = df.iloc[:, 1:]
y
```

```python
df.corr(numeric_only = True)
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
#Model performace

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,mean_absolute_percentage_error,root_mean_squared_error
mae=mean_absolute_error(y,y_pred)
mae
```

```python
mape = mean_absolute_percentage_error(y,y_pred)
mape
```

```python
mse = mean_squared_error(y,y_pred)
mse
```

```python
rmse = root_mean_squared_error(y,y_pred)
rmse
```

```python
r2_sc = r2_score(y,y_pred)
r2_sc
```

```python
from sklearn.preprocessing import PolynomialFeatures
feature = PolynomialFeatures(degree=2)
x_poly = feature.fit_transform(x)
x_poly
```

```python
model1 = LinearRegression() #We already imported linear regression
model1.fit(x_poly,y)
y_poly = model1.predict(x_poly)
y_poly
```

```python
y
```

```python
mape1 = mean_absolute_percentage_error(y,y_poly)
mape1
```

```python
r2_sc1= r2_score(y,y_poly)
r2_sc1
```

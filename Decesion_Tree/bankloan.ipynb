```python
#Drop Unnecessary Columns
#we can drop the coulumns id and zip code as they'r not useful
df.drop(['ID','ZIP.Code'],inplace=True,axis=1)
```

```python
X = df.drop(columns=['Personal.Loan'])
X
```

```python
Y = df['Personal.Loan']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Model creation
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
y_test #actual value
```

```python
#Draw the tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(model)
```

```python
#As the x[3],x-2,x[1] not know ,so we need to mention that.
#filled = True [To add the colors]

feature_cols = ['Age','Experience','Income','Family','CCAvg','Education',
                'Mortgage','Securities.Account','CD.Account','Online','CreditCard']
tree.plot_tree(
    model,
    feature_names=feature_cols,
    class_names=['No Loan', 'Loan'],
    filled=True
)
```

```python
#Performance Evaluation method

from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred) #Comparing values, y_test:actual value and y-pred:predicted value
cm
```

```python
#Accurace score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score 
```

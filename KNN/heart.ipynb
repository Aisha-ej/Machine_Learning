```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test) 
y_pred
```

```python
y_test
```

```python
#create confusion matrics first
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred) #Comparing values, y_test:actual value and y-pred:predicted value
cm
```

```python
#confusion matrics display
from sklearn.metrics import ConfusionMatrixDisplay
lab=['0', '1'] #We need to give the class labels as the same y_train
cmd=ConfusionMatrixDisplay(cm,display_labels=lab) #created object for confusion matrics. cm:confusionmatrics,display_label=lab
cmd.plot()
```

```python
#Accurace score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score 
```

```python
#Hyper Parameter Tunning
from sklearn.model_selection import GridSearchCV
model1 = KNeighborsClassifier()
data = {'n_neighbors': [3, 5, 7, 9, 11],'weights': ['uniform', 'distance']}
grid = GridSearchCV(model1,data,cv=10,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_) #To find which parameter gives the best performace
```

```python
print(grid.best_score_)
```

```python
model2 = KNeighborsClassifier(n_neighbors=9,weights='distance')
model2.fit(x_train,y_train)
y_pred1 = model2.predict(x_test)
y_pred
```

```python
score =accuracy_score(y_test,y_pred1)
score #Increased the performace
```

```python
x = df.iloc[:,:-1].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
scaler.fit(x_train) #collecting the x_train to find the mean of training data
x_train=scaler.transform(x_train) #x_train transformed to normalized form
x_test=scaler.transform(x_test) #y_train transformed to normalized form
x_train
```

```python
#Create Machine Leanining Algorithm

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test) #machine predict the values according to what we lean teached to machine
y_pred
```

```python
y_test
```

```python
#Accurace score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score #Its a good model
```

```python
#As the perfomrace is below 70,its not good performace, we can increase the perofrmance using Hyper Parameter Tunning
```

```python
#Hyper Parameter Tunning

#Many hyper parameter methods are there. Here,we're using the below one.

#GridSearch CV
#Random Search
#Bayesian Optimization
#Here we're choosing two parameters. n-neighbours and wieght. machine will tells that will give maximum performace with one of these parameter.

```

```python
#GridSearch CV
from sklearn.model_selection import GridSearchCV
model1 =KNeighborsClassifier()
data ={
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance']
}  #The parameters we're going to tune here
grid = GridSearchCV(model1,data,cv=10,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_) #To find which parameter gives the best performace
```

```python
print(grid.best_score_)
```

```python
model2 = KNeighborsClassifier(n_neighbors=7,weights='distance')
model2.fit(x_train,y_train)
y_pred1 = model2.predict(x_test)
y_pred1
```

```python
score =accuracy_score(y_test,y_pred1)
score #Here performace has decreases as we took 2 paremeters here. so there is a chance of improving or decreassing the perofrmace while doing Hyper Paramter Tunning.
```

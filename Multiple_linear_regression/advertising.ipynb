```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('advertising.csv')
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#Correlation

df.corr(numeric_only=True)
```

```python
#Seaborn
#regplot
#As seaborn is an advanced visualization methos, we an plot scatter plot and regeression line in one single step using regplot
```

```python
#TV VS Sales
sns.regplot(x = df['TV'],y = y)
```

```python
#Radio VS Sales
sns.regplot(x = df['Radio'],y = y)
```

```python
#Newspaper VS Sales
sns.regplot(x = df['Newspaper'],y = y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Model Creation
#Simple Linear Regression Model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
x_test
```

```python
y_test
```

```python
#print(model.predict([[6.8]]))
```

```python
#Data Frame

df1 = pd.DataFrame({'Actual Value': y_test,'Predicted Value': y_pred,'Difference':y_test-y_pred})
df1
```

```python
#print the slop
print(model.coef_) #Coefficient .Another name for slop
```

```python
#Constant
print(model.intercept_)
```

```python
#Model performace

from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test,y_pred)
mape 
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_test,y_pred)
rmse
```

```python
from sklearn.metrics import r2_score
r2_sc = r2_score(y_test,y_pred)
r2_sc
```
